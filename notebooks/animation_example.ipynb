```python
import numpy as np
import matplotlib.pyplot as plt

# import seaborn and set a output style
import seaborn as sns
sns.set_theme(style="darkgrid")

# create 10000 samples using normal low N(0, 2)
samples = np.random.normal(loc=0, scale=2.0, size=10000)

# display these samples using matplotlib and seaborn
fig = plt.figure(figsize=(15, 10))
plt.hist(samples, bins=50)
plt.title('Simple normal distribution with N(mean=0, std=2.0)', fontsize=20)
plt.show()
```

```python
import matplotlib.animation as animation
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# set seaborn darkgrid theme
sns.set_theme(style="darkgrid")

# create samples distribution with different `std` values
samples_study = [ np.random.normal(loc=0, scale=i, size=10000) for i in range(2, 12) ]

# generate expected titles
titles = [ f"Frame nÂ°{index + 1} -- N(0, {std})" for index, std in enumerate(range(2, 12)) ]

# func which will be applied in order to create a video frame
def func(frame, ax, current_samples, titles):
    
    ax.clear()
    # density will displayed value in an normalized way
    ax.hist(current_samples[frame], bins=50, color='steelblue', density=True)
    ax.set_xlim(-30, 30)
    ax.set_ylim(0, 0.25) # set y_lim following density normalized
    ax.set_title(titles[frame], fontsize=20)

    return ax

fig = plt.figure(figsize=(15, 10))

# get number of expected frames
frames_indices = range(len(titles))

# get the subplot to update when generating the animation
ax = fig.add_subplot(111)

# create the animation using custom func and its parameters (fargs)
ani = animation.FuncAnimation(fig, func, frames_indices, interval=1000,
                              repeat_delay=1000, blit=False,
                              fargs=(ax, samples_study, titles))

# save the animation into a specific video format
ani.save('./images/animation_distribution.gif')

#writergif = animation.PillowWriter(fps=5)
#ani.save('filename.gif',writer=writergif)
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import math

x = np.linspace(0,2*math.pi,100)
y = np.sin(x)    

fig, ax = plt.subplots()
line, = ax.plot([])

ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.1, 1.1)

def animate(frame_num):
    y = np.sin(x + 2*np.pi * frame_num/100)
    line.set_data((x, y))
    return line

anim = FuncAnimation(fig, animate, frames=40, interval=1)

anim.save('./images/sine.gif')
#writergif = animation.PillowWriter(fps=30)
#anim.save('sine.gif', writer=writergif)
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import math

x = np.linspace(0,2*math.pi,100)
X = np.tile(x, (100, 1)) 
    
fig, ax = plt.subplots()

def animate(frame_num):
    Z = np.sin(X + 2*np.pi * frame_num/100)
    ax.imshow(Z)
    return ax

anim = FuncAnimation(fig, animate, frames=40, interval=1)

anim.save('./images/sine3d.gif')
```
