```python
import sys
sys.path.insert(0, '../../')
import math
import numpy as np
import matplotlib.pyplot as plt
from gasp import ssfp, responses, simulation, dataset
```

```python
Mc = dataset.load_dataset8c()
print(Mc.shape)
```

```python
Mc0 = Mc[:,:,:,0]
Mc1 = Mc[:,:,:,1]
Mc2 = Mc[:,:,:,2]

field_map = np.angle(np.conjugate(Mc0) * Mc2)
f = np.exp(-1j * field_map / 2)
Mw = Mc0 + Mc1
Mf = Mc0 - Mc1
Mw3 = Mc0 + Mc1 * np.exp(-1j * field_map / 2)
Mf3 = Mc0 - Mc1 * np.exp(-1j * field_map / 2)

Mw = np.sqrt(np.sum(np.abs(Mw)**2, axis=2))
Mf = np.sqrt(np.sum(np.abs(Mf)**2, axis=2))
Mw3 = np.sqrt(np.sum(np.abs(Mw3)**2, axis=2))
Mf3 = np.sqrt(np.sum(np.abs(Mf3)**2, axis=2))
```

```python
f = plt.figure(figsize=(8,8))
ax = f.add_subplot(2, 2, 1)
ax2 = f.add_subplot(2, 2, 2)
ax.imshow(np.abs(Mw), cmap='gray')
ax2.imshow(np.abs(Mf), cmap='gray')
plt.show()
```
